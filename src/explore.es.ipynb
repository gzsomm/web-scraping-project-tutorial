```python
# Seleccionar el recurso a descargar
resource_url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Petición para descargar el fichero de Internet
response = requests.get(resource_url, time.sleep(10), headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36"})

# Si la petición se ha ejecutado correctamente (código 200), entonces el contenido HTML de la página se ha podido descargar
if response.status_code == 200:
    # Transformamos el HTML plano en un HTML real (estructurado y anidado, con forma de árbol)
    soup = BeautifulSoup(response.text, 'html')
    print(soup)
else:
    print (f"Error {response.status_code}")
```

```python
tablas = soup.find_all("table")
print(tablas)
```

```python
for i in tablas:
  print(i.text)
```

```python
import pandas as pd
```

```python
for index, tablas in enumerate (tablas):
    if ("Tesla Quarterly Revenue" in str(tablas)):
        tabla_objetivo = index
        break

print (tabla_objetivo)
```

```python
# Create a DataFrame
tesla_revenue2 = pd.DataFrame(columns = ["revenue_date", "revenue"])
print (tesla_revenue2)

print(tablas)


```

```python
for row in tablas[tabla_objetivo].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        revenue_date = col[0].text
        revenue = col[1].text.replace("$", "").replace(",", "")

        tesla_revenue2 = pd.concat([tesla_revenue2, pd.DataFrame({
            "revenue_date": revenue_date,
            "revenue": revenue
        }, index = [0])], ignore_index = True)

tesla_revenue2.head()
```

```python
con = sqlite3.connect("tesla.db")
```

```python
cursor = con.cursor()
cursor.execute("DROP TABLE IF EXISTS Quarterly_Revenue")
```

```python
con.execute("""CREATE TABLE Quarterly_Revenue (
    revenue_date DATE,
    revenue INT
)""")
```

```python
tesla_revenue2.to_sql('Quarterly_Revenue', con, if_exists='append', index=False)
```
